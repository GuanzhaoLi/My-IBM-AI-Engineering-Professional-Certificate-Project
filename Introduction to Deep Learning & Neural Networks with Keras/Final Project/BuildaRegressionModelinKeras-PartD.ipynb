```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
csv_data = pd.read_csv('concrete_data.csv')
csv_data.describe()
```

```python
other_data = csv_data[['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age']]
strength_data = csv_data['Strength']
n_cols = other_data.shape[1]
print(n_cols)
```

```python
other_data.head()
```

```python
#column_other = ['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age']

for item in other_data.columns:
    data = other_data[item]
    new_data = (data - np.mean(data))/np.std(data)
    #print(new_data)
    other_data[item]= new_data

data = strength_data
new_data = (data - np.mean(data))/np.std(data)
strength_data= new_data

print(strength_data)
```

```python
def my_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = my_model()
msr = []

for i in range(50):
    x_train, x_test, y_train, y_test = train_test_split(other_data, strength_data, test_size=0.3, random_state=42)
    model.fit(x_train, y_train, epochs = 100, verbose= 1)
    y_predict = model.predict(x_test)
    msr.append(mean_squared_error(y_test, y_predict))
```

```python
arr = np.array(msr)
print(np.mean(arr))
print(np.std(arr))
```
